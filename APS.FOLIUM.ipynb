```python
import folium
import pandas as pd
```

```python
map = folium.Map(
    zoom_start=5,
    location=[-15.159003511092683, -48.0894289512393]
)
```

```python
ubs = pd.read_csv('Unidades_Basicas_Saude-UBS.csv', sep=';', decimal=',')
ubs
```

```python
for i in range(0, 50):
    unidade = ubs.loc[i]
    folium.Marker(
        location=[unidade['LATITUDE'], unidade['LONGITUDE']]
    ).add_to(map)
map
```
